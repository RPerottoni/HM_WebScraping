```python
import re
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
# Classic Category Books: Data Colect

# API Request
url = 'https://books.toscrape.com/catalogue/category/books/classics_6/index.html'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}
page = requests.get( url, headers=headers)
soup = BeautifulSoup(page.text, 'html.parser')

# Beautiful Soup Object
soup = BeautifulSoup(page.text, 'html.parser')

# Classic books data colect
classic_products = soup.find('div', class_='col-sm-8 col-md-9')

# ===== Books Name =====
classic_book = classic_products.find_all('h3')
classic_bookname = [p.get_text('title') for p in classic_book]
df_book = pd.DataFrame(classic_bookname)
df_book.columns = ['Book Name']

# ===== Books Rating =====
classic_rating = classic_products.find_all('p', class_='star-rating')
classic_rating = str(classic_rating)
regex = '([A-Z].+)\"'
df_rating = pd.DataFrame(re.findall( regex, classic_rating))
df_rating.columns=['Book Rating']

# ===== Books Availability =====
availability_bookstock = classic_products.find_all('p', class_='instock availability')
availability_book = [p.get_text('instock availability') for p in availability_bookstock]
availability_book = str(availability_book)
regex = '([A-Z]\w [a-z]+k)'
df_availability = pd.DataFrame(re.findall(regex, availability_book))
df_availability.columns = ['Availability']

# ===== Books Price =====
classic_price = classic_products.find_all('p', class_='price_color')
product_price = [p.get_text('price_color') for p in classic_price]
df_price = pd.DataFrame(product_price)
df_price.columns = ['Book Price']
df_price['Book Price'] = df_price['Book Price'].apply( lambda x: x.replace( 'Â£', ''))

# Empty DataFrame and add colected data on dataframe.
df_classicbooks = pd.concat([df_book, df_rating, df_price, df_availability], axis = 1)

# Add Scrapy time and Books Category
df_classicbooks['Category'] = 'Classic'
df_classicbooks['Scrapy Time'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

# Print the DataFrame
df_classicbooks
```
